## Total Fares by City Type

```python
# Calculate Type Percents
urban_fares = df_merge.loc[df_merge['type'] == 'Urban', 'fare'].sum()
suburb_fares = df_merge.loc[df_merge['type'] == 'Suburban', 'fare'].sum()
rural_fares = df_merge.loc[df_merge['type'] == 'Rural', 'fare'].sum()
total_fares = urban_fares + suburb_fares + rural_fares

city_types = ['Urban', 'Suburban', 'Rural']
city_fares = [urban_fares, suburb_fares, rural_fares]
type_colors = ['gold', 'lightcoral', 'lightblue']
ride_explode = (0, 0.1, 0.2)

plt.pie(city_fares, explode=ride_explode, labels=city_types, colors=type_colors, autopct="%1.2f%%", shadow=True, startangle=220)
plt.title("% of Total Fares by City Type")
plt.show()
# Build Pie Chart

# Save Figure
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
type_totals = df_merge['type'].value_counts()
total = 1625 + 625 + 125

urban_total = 1625 / total * 100
suburb_total = 625 / total * 100
rural_total = 125 / total * 100

city_types = ['Urban', 'Suburban', 'Rural']
type_pcts = [urban_total, suburb_total, rural_total]
type_colors = ['gold', 'lightcoral', 'lightblue']
ride_explode = (0, 0.1, 0.1)

plt.pie(type_pcts, explode=ride_explode, labels=city_types, colors=type_colors, autopct="%1.1f%%", shadow=True, startangle=220)
plt.title('% of Total Rides by City Type')
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_drivers = df_merge.loc[df_merge['type'] == 'Urban', 'driver_count'].sum()
suburb_drivers = df_merge.loc[df_merge['type'] == 'Suburban', 'driver_count'].sum()
rural_drivers = df_merge.loc[df_merge['type'] == 'Rural', 'driver_count'].sum()

total_drivers = urban_drivers + suburb_drivers + rural_drivers

pct_urban_drivers = urban_drivers / total_drivers * 100
pct_suburb_drivers = suburb_drivers / total_drivers * 100
pct_rural_drivers = rural_drivers / total_drivers * 100

# Build Pie Charts
city_types = ['Urban', 'Suburban', 'Rural']
type_pcts = [pct_urban_drivers, pct_suburb_drivers, pct_rural_drivers]
type_colors = ['gold', 'lightcoral', 'lightblue']
driver_explode = (0, 0.1, 0.1)

plt.pie(type_pcts, explode=driver_explode, labels=city_types, colors=type_colors, autopct="%1.2f%%", shadow=True, startangle=180)
plt.title('% of Total Drivers by City Type')
plt.show()

# Save Figure
```
